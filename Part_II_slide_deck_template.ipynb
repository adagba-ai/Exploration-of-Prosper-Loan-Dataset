## Dataset Overview

> Prosper loan dataset contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loan_df = pd.read_csv('loan_df.csv')
loan_df
```

```python
def type_counts(x):
     return x.value_counts()
```

## (Visualization 1)

> The Bar Chart below shows the Number of recommendations (1) and no-recommendations(0) borrowers had at the time the listing was created with the aim of  determining if a large number of recommendations signify a higher chance of getting a Prosper loan. However, 0 accounted for the highest number of Recommendationhis. This signifies that having a high number of Recommendation does not guarantee getting a loan. Other factors could have been considered.



```python
loan_df['Recommendations'].value_counts().plot(kind='bar', figsize=[12, 8]).set_ylabel('Number of recommendations')
plt.xlabel('Recommendations')
plt.legend()
plt.title('Bar Chart showing Number of recommendations');
```

## (Visualization 2)

> The following chart depict how possible it is that Prosper could have considered other factors in determinig the suitability of a Borrower. The Bar Chart shows that over 60,000 Borrowers are employed. This is a good indication that Proper consider mostly employed people for their loan. Its common to note that their reason for this could easily be that only people with jobs can repay their loan.

```python
plt.figure(figsize=[12, 8])
freq = type_counts(loan_df['EmploymentStatus'])
base_color = sb.color_palette()[0]
gen_order = freq.index
plt.title('Decreasing order of the frequency of the Employment Status')
sb.countplot(data=loan_df, x='EmploymentStatus', color=base_color, order=gen_order);
plt.legend(labels=["EmploymentStatus"])
plt.xticks(rotation=90);
```

## (Visualization 3)

> To further stress this fact, the chart below illustrate how Borrowers whose income is verifiable stands a better chance of getting Prosper loan compared to those who are not verified.

```python
plt.figure(figsize=[12,8])
type_counts(loan_df['IncomeVerifiable']).plot(kind = 'bar')
plt.title('Borrowers Income Verifiable')
plt.ylabel('Count')
plt.xlabel('Borrower Income Verifiable Classification')
plt.legend(fontsize=15)
plt.show()
```

## (Visualization 4)

> Lets look at What category of Employment Status had their income verified the most. From the visual below, we can see that Borrowers who are Not-employed had their income verified the most. Followed by those who are self-employed. We can also see that for Borrowers who are employed there is little verification of income. These analysis goes further to justify our earlier claim that Prosper are more interested Borrowers who are employed, hence will definately verify the income of those who are not employed.

```python
sb.catplot(
    x='EmploymentStatus', 
    y='Recommendations', 
    data=loan_df,
    palette='bright',
    kind='bar',
    hue='IncomeVerifiable', height=8, aspect=12/8)
plt.title('Rcommendation of Borrowers Income Verifiable')
plt.xticks(rotation = 90);
```

## (Visualization 5)
> In the next chart, we are interested in what Category of Occupation had a better Chance of Getting Prosper Loan, and the chart here indicated that Others in the Category of Occupation had a better chance of getting Prosper Loan with a count or frequency of 28000. 
> Another group is Proffessional with a frequency of 14000. Computer Programmer-Lucky we, had a frequncy of 4000,  
showing a good chance to get prosper loan as a Computer Programmer. The Chart shows that Student had the least chance of getting a loan from Prosper.

```python
freq = loan_df['Occupation'].value_counts()
plt.figure(figsize=(12,8))
gen_order = freq.index
plt.title('Occupation of Borrowers')
sb.countplot(data=loan_df, x='Occupation', color=base_color, order=gen_order);
plt.legend(labels=["Occupation"])
plt.xticks(rotation=90);
```
